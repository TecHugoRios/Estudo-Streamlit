```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use('_mpl-gallery')
df = pd.read_excel("plastic-waste-generation-2000-2019.xlsx")

plasticoMundo = df.loc[df['Entity'] == 'World']
plasticoMundo = pd.DataFrame(plasticoMundo)

display(plasticoMundo)
display(plasticoMundo['Year'])
display(plasticoMundo["Total waste"])
```

```python
x = plasticoMundo["Year"]
y = np.array(plasticoMundo["Total waste"])

plt.figure(figsize=(10,6))
plt.plot(x,y, color='blue')
plt.xticks(ticks=plasticoMundo['Year'], labels=plasticoMundo['Year'].astype(int))
plt.scatter(x,y, color='red')
plt.title("Total de Plástico no Mundo")
plt.xlabel("Anos")
plt.ylabel("Desperdício Total (toneladas)")
plt.legend(['Estimativa', 'Dado'])

#Adicionar texto dos valores de cada ponto no gráfico
'''for i in range(len(plasticoMundo)):
    plt.text(plasticoMundo['Year'][i], plasticoMundo['Total waste'][i] + 2, f'{plasticoMundo["Total waste"][i]}', 
             fontsize=10, ha='right', color='darkblue')'''

plt.tight_layout() #Ajusta o layout para evitar sobreposição
plt.show()
```
